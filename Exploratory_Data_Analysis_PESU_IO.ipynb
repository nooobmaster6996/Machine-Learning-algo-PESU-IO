# Data Visualisation

```python
fig=px.histogram(dataset,'Age')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig=px.histogram(dataset,'AnnualIncome')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
px.pie(dataset,'Employment Type',title='Distribution of Employment type of Individuals')
```

```python
px.pie(dataset,'GraduateOrNot',labels={'GraduateOrNot':'Graduated'},title='Distribution of graduated individuals')
```
